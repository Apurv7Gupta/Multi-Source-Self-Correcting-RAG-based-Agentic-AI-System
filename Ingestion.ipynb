```python
import os
from typing import List
from langchain_community.document_loaders import UnstructuredFileLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
import chromadb
```

```python
# --- 1. CONFIGURATION ---
CHROMA_HOST = os.getenv("chroma_server", "localhost")
CHROMA_PORT = os.getenv("chroma_port", "8000")
COLLECTION_NAME = "enterprise_knowledge"
```

```python
# --- 2. INITIALIZE EMBEDDINGS ---
# Consistent with the RAG node to ensure vector space alignment
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

client = chromadb.HttpClient(host=CHROMA_HOST, port=CHROMA_PORT)

vector_db = Chroma(
    collection_name="collection1",
    embedding_function=embeddings,
    client=client
)
```

```python
# --- 3. CORE INGESTION LOGIC ---


from langchain_community.document_loaders import UnstructuredURLLoader
from langchain_community.tools.tavily_search import TavilyAnswer

def ingest_web_url(url: str):
    """
    Scrapes a live URL and adds it to the vector store.
    """
    loader = UnstructuredURLLoader(urls=[url])
    docs = loader.load()
    
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
    chunks = text_splitter.split_documents(docs)
    
    # Use your defined chroma_client to sync
    vector_db.add_documents(chunks)
    return f"URL {url} ingested successfully."

# ADDING THE SEARCH TOOL FOR THE AGENT
from langchain_community.tools.tavily_search import TavilySearchResults

# Industry standard for agentic search
search_tool = TavilySearchResults(k=3)


def process_document(file_path: str):
    """
    Handles PDF/Markdown/Docx ingestion using Unstructured.
    Professional-grade partitioning for complex layouts.
    """
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Source file not found: {file_path}")

    # Use Unstructured to handle layout-aware partitioning
    # 'elements' mode preserves hierarchy better than 'single'
    loader = UnstructuredFileLoader(
        file_path,
        strategy="fast",  # Use "hi_res" if you need OCR for scanned PDFs
        mode="elements",
    )

    raw_docs = loader.load()

    # Recursive splitting is the industry standard for maintaining context
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000, chunk_overlap=150, add_start_index=True
    )

    chunks = text_splitter.split_documents(raw_docs)

    # Add metadata for search filtering (Enterprise standard)
    for chunk in chunks:
        chunk.metadata["source"] = os.path.basename(file_path)

    return chunks


def sync_to_vector_db(chunks: List):
    """
    Syncs processed chunks to a remote Vector DB instance.
    """
    # Connect to a persistent Chroma instance (Render-compatible)
    client = chromadb.HttpClient(host=CHROMA_HOST, port=CHROMA_PORT)

    vectorstore = Chroma(
        client=client, collection_name=COLLECTION_NAME, embedding_function=embeddings
    )

    # Batch upsert to prevent API timeouts on large documents
    vectorstore.add_documents(chunks)
    print(f"Successfully ingested {len(chunks)} chunks into {COLLECTION_NAME}.")
```

```python
# --- 4. EXECUTION ---

if __name__ == "__main__":
    # In a real-world scenario, this is triggered by a webhook or S3 event
    target_file = "data/knowledge_base.pdf"

    try:
        doc_chunks = process_document(target_file)
        sync_to_vector_db(doc_chunks)
    except Exception as e:
        print(f"Ingestion failed: {str(e)}")
```
